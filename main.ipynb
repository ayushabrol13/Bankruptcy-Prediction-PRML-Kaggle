```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('data/train')
df_test = pd.read_csv('data/test')
```

```python
df_train.drop(["Attr5"], axis=1, inplace = True)
df_train
```

```python
df_test.drop(["Attr5"], axis=1, inplace = True)
df_test
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
X, y = df_train.iloc[:, :-1], df_train.iloc[:, -1]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
values, counts = np.unique(y_train, return_counts=True)
print(values, counts)
```

```python
import xgboost as xgb
# Firstly, I tuned the hyperparameters and then used them for the maximum roc_auc_score in xgboost.
xg_reg = xgb.XGBClassifier(objective="binary:logistic", eval_metric = "auc", subsample = 0.9, max_depth = 6, n_estimators = 115, learning_rate=0.2, n_jobs=-1)
xg_reg.fit(X_train, y_train)
y_pred_xg = xg_reg.predict_proba(X_val)
print(roc_auc_score(y_val, y_pred_xg[:,1]))
```

```python
y_pred_xg
```

```python
y_pred_test = xg_reg.predict_proba(df_test)
y_pred_test = y_pred_test[:,1]
```

```python
import lightgbm as lgb
# Firstly, I tuned the hyperparameters and then used them for the maximum roc_auc_score in lightgbms.
lg_reg = lgb.LGBMClassifier(objective="binary", subsample = 1, max_depth = 6, n_estimators = 100, learning_rate=0.1, n_jobs=-1)
lg_reg.fit(X_train, y_train)
y_pred_lg = lg_reg.predict_proba(X_val)
print(roc_auc_score(y_val, y_pred_lg[:,1]))
```

```python
from sklearn.ensemble import VotingClassifier
# Then, I used a voting classifier to select the best gradient boost ensemble using the weights. Best fit was for 2:1 . 
model = VotingClassifier(estimators=[('xg', xg_reg), ('lg', lg_reg)], voting='soft', weights=[2,1])
model.fit(X, y)
y_pred_voting = model.predict_proba(df_test)
# y_pred_voting = model.predict_proba(X_val)
# print(roc_auc_score(y_val, y_pred_voting[:,1]))
```

```python
indexes = [int(i) for i in range(0, len(y_pred_voting[:, 1]))]
```

```python
# Creating the final csv.
pd.DataFrame({'Id': indexes, 'predicted': y_pred_voting[:, 1]}).to_csv('submission.csv', index=False)
```
